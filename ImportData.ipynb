## Import file location

```python
example2 = "dataset/Lending-Company-Numeric-Data.csv"
example3 = "dataset/Lending-Company-Numeric-Data-NAN.csv"
```

## Reading data with numpy

```python
import numpy as np
```

## Load txt 

#### Load txt will throw an error if the data that loaded in is contains a null or empty data

```python
example2_loadtxt = np.loadtxt(example2, delimiter=',')
example2_loadtxt
```

## Gen from txt

#### Gen from txt will fill the empty or null data by empty string " " if the data that loaded in is empty

```python
example3_gentxt = np.genfromtxt(example3, delimiter=';')
example3_gentxt
```

```python
example3_gentxt_partial = np.genfromtxt(example3, delimiter=';', skip_header=2,skip_footer=2, usecols=(5,0,1))
example3_gentxt_partial
```

```python
set_example3_setdtype = np.genfromtxt(example3, delimiter=';', dtype = "str", skip_header=2,skip_footer=2, usecols=(5,0,1))
set_example3_setdtype
```

### Save data to json

#### Data can be saved using a `savetxt` and fill the required parameter like `fmt=` format of the file exported, `delimiter=` for the delimiter if its separator
> For the extension you can explicitly write in the exported filename in the first parameter

```python
np.savetxt('example.json', set_example3_setdtype, fmt='%s', delimiter=',') 
```

```python
np.savetxt('example.xlsx', set_example3_setdtype, fmt='%s', delimiter=',')
```
