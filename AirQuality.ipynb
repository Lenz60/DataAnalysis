```python
import pandas as pd
import numpy as np
```

```python
df_jakarta_full = pd.read_csv('Dataset/DKI_Full.csv')
df_jakarta_HI = pd.read_csv('Dataset/DKI_HI.csv')
df_jogja_21_jan = pd.read_csv('Dataset/DIY_Jan_2021.csv')
```

## Data Manipulation on the Jogjakarta dataset 


> Filter the data by sorting it to the only September 2021
> <br>
> Filter the data by using time on 9AM only each date


```python
df_jogja_21_jan

# Convert 'Date' column to datetime type
df_jogja_21_jan['Date'] = pd.to_datetime(df_jogja_21_jan['Date'])

# Filter the data frame for September 2021
df_january = df_jogja_21_jan[df_jogja_21_jan['Date'].dt.month == 1]

# Filter the data frame for time 09:00:00
df_jogja_jan_21 = df_january[df_january['Time'] == '09:00:00']

# Print the filtered data frame
df_jogja_jan_21
```

> Reset the index


```python
df_jogja_jan_21.reset_index(drop=True, inplace=True)
df_jogja_jan_21
```

```python
# df_jogja_jan_21['Category'] = df_jogja_jan_21['Category'].replace({'Good': 'Baik', 'Moderate': 'Sedang'})
# df_jogja_jan_21
```

> Filling the NaN data to 0

```python
df_jogja_jan_21.fillna(0, inplace=True)
df_jogja_jan_21
```

> Delete the coloumn of Time

```python
df_jogja_jan_21 = df_jogja_jan_21.drop(columns=['Time'])
df_jogja_jan_21
```

> Added City column


```python
df_jogja_jan_21['City'] = 'Yogyakarta'
df_jogja_jan_21
```

# Manipulate Data Jakarta HI

```python
df_jakarta_HI
```

> Filter the data to the January 2021


```python
# Convert the "tanggal" column to datetime format
df_jakarta_HI['tanggal'] = pd.to_datetime(df_jakarta_HI['tanggal'])

# Filter the data to include only September 2021
df_jakarta_jan_21 = df_jakarta_HI[(df_jakarta_HI['tanggal'].dt.month == 1) & (df_jakarta_HI['tanggal'].dt.year == 2021)]

# Display the filtered data
df_jakarta_jan_21
```

> Reset the index

```python
df_jakarta_jan_21.reset_index(drop=True, inplace=True)
df_jakarta_jan_21
```

> Rename the status to english

```python
df_jakarta_jan_21['categori'] = df_jakarta_jan_21['categori'].replace({'BAIK': 'Good', 'SEDANG': 'Moderate', 'TIDAK SEHAT': 'Worst'})
df_jakarta_jan_21
```

> Remove stasiun column

```python
df_jakarta_jan_21 = df_jakarta_jan_21.drop(columns=['stasiun'])
```

> Rename the Header

```python
# Date	PM10	PM2.5	SO2	CO	O3	NO2	Max	Critical Component	Category
df_jakarta_jan_21 = df_jakarta_jan_21.rename(columns={'tanggal': 'Date', 
                                                        'pm10': 'PM10', 
                                                        'pm25': 'PM2.5', 'so2': 'SO2', 
                                                        'co': 'CO', 'o3': 'O3', 
                                                        'no2': 'NO2',
                                                        'critical': 'Critical Component', 
                                                        'max': 'Max', 'categori': 'Category'})
```

```python
df_jakarta_jan_21
```

> Remove the decimals


```python
df_jakarta_jan_21 = df_jakarta_jan_21.astype(str).replace("\.0", "", regex=True)
```

```python
df_jakarta_jan_21
```

> Re index the header and data

```python
# Date	PM10	PM2.5	SO2	CO	O3	NO2	Max	Critical Component	Category
df_jakarta_jan_21 = df_jakarta_jan_21.reindex(columns=['Date', 'PM10', 'PM2.5', 'SO2', 'CO', 'O3', 'NO2', 'Max', 'Critical Component', 'Category'])
```

```python
df_jakarta_jan_21
```

> Added City column


```python
df_jakarta_jan_21['City'] = 'Jakarta'
```

```python
df_jakarta_jan_21
```

# Clean data Jakarta Full


```python
df_jakarta_full
```

```python
df_jakarta_full['categori'] = df_jakarta_full['categori'].replace({'BAIK': 'Good', 'SEDANG': 'Moderate', 'TIDAK SEHAT': 'Worst', 'TIDAK ADA DATA': 'No Data'})
df_jakarta_clean = df_jakarta_full
df_jakarta_clean
```

```python
df_jakarta_clean = df_jakarta_clean.rename(columns={'tanggal': 'Date',
                                                        'stasiun': 'Scope',
                                                        'pm10': 'PM10', 
                                                        'pm25': 'PM2.5', 'so2': 'SO2', 
                                                        'co': 'CO', 'o3': 'O3', 
                                                        'no2': 'NO2',
                                                        'critical': 'Critical Component', 
                                                        'max': 'Max', 'categori': 'Category'})
```

```python
df_jakarta_clean
```

```python
df_jakarta_clean['PM2.5'].fillna(df_jakarta_clean['PM2.5'].mean().round(), inplace=True)
df_jakarta_clean
```

```python
df_jakarta_clean = df_jakarta_clean.astype(str).replace("\.0", "", regex=True)
df_jakarta_clean
```

```python
df_jakarta_clean = df_jakarta_clean.reindex(columns=['Date', 'Scope','PM10', 'PM2.5', 'SO2', 'CO', 'O3', 'NO2', 'Max', 'Critical Component', 'Category'])
df_jakarta_clean
```

```python
df_jakarta_clean['Scope'] = df_jakarta_clean['Scope'].str[5:]
df_jakarta_clean
```

```python
df_jakarta_clean['Scope'] = df_jakarta_clean['Scope'].str.replace(r'[()]', '', regex=True)
df_jakarta_clean
```
