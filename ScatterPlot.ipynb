```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
sns.set_theme()
```

```python
df_real_estate = pd.read_csv('dataset/scatter_data.csv')
```

```python
plt.figure(figsize=(12, 8))
scatter = plt.scatter(df_real_estate['Area (ft.)'], df_real_estate['Price'],  
            alpha=0.6, 
            c=df_real_estate['Building Type'],
            cmap='viridis',
            edgecolors='grey',)

plt.legend(*scatter.legend_elements(),
           loc = 'lower right', 
           title ='Building Type')

plt.title('Relationship between Area and Price of California Real Estate', 
          fontsize=16,
          weight='bold')

plt.xlabel('Area (sq. ft.)')
plt.ylabel('Price ($)')

plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x = df_real_estate['Area (ft.)'], 
                y = df_real_estate['Price'],  
                alpha=0.6, 
                hue=df_real_estate['Building Type'],
                palette='viridis',
                s=80,)

plt.legend(*scatter.legend_elements(),
           loc = 'lower right', 
           title ='Building Type')

plt.title('Relationship between Area and Price of California Real Estate', 
          fontsize=16,
          weight='bold')

plt.xlabel('Area (sq. ft.)')
plt.ylabel('Price ($)')

plt.show()
```
