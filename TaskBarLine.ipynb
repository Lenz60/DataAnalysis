```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
import seaborn as sns
sns.set_theme()
```

```python
df_data = pd.read_csv('dataset/bar_line_chart_data.csv')
```

```python
df_data
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(df_data['Year'], df_data['Participants'], color='k')
ax.set_ylabel('Number of Participants',
              weight = 'bold')

ax.tick_params(axis ='y', 
               colors ='k',
               width = 2,
               labelsize = 'large')

ax1 = ax.twinx()
ax1.set_ylim(0,1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax=1))
ax1.plot(df_data['Year'], df_data['Python Users'],
         color = 'r', marker = 'D')
ax1.set_ylabel('Python Users',
              color = '#b60000',
            weight = 'bold')
ax1.tick_params(axis ='y',
                colors = '#b60000')
```

```python
df_data2 = pd.read_csv('dataset/bar_line_chart_homework.csv')
```

```python
df_data2
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title('Movie Theater Goers Complaints vs Frequency')

bar = ax.bar(df_data2['Movie Theater Goers Complaints'], 
       df_data2['Number of complaints'], 
       color='mediumseagreen',
       edgecolor='black',
       alpha=0.7)

ax.set_ylabel('Number of Complaints',
              weight = 'bold')

ax.tick_params(axis ='y', 
               colors ='k',
               width = 2,
               labelsize = 'large')

bar[0].set_color('seagreen')
bar[0].set_edgecolor('black')
ax1 = ax.twinx()
ax1.set_ylim(0,1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax=1))
ax1.plot(df_data2['Movie Theater Goers Complaints'], df_data2['frequency'],
         color = 'r', marker = 'D')
ax1.set_ylabel('Frequency (%)',
              color = '#b60000',
            weight = 'bold')
ax1.tick_params(axis ='y',
                colors = '#b60000')
```
