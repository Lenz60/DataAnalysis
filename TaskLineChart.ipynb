```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
df_data = pd.read_csv('dataset/line_chart_homework_data.csv')
```

```python
df_data
```

```python
df_data['New_date'] = pd.to_datetime(df_data['Date'])
```

```python
df_data
```

## Overall Line Chart Data

```python
# Set the laels for the legend
labels = ['S&P Ret', 'FTSE Ret']
plt.figure(figsize=(20, 8))

# Create a line plot
plt.plot(df_data['New_date'], 
         df_data['GSPCRet'], 
         color='forestgreen', label='GSPCRet')

plt.plot(df_data['New_date'], 
         df_data['FTSERet'], 
         color='royalblue', label='FTSERet')

# Set title, labels and legend
plt.title('S&P Ret and FTSE Ret Index ( 2010 - 2020)', fontsize=15, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Returns', fontsize=12)
plt.legend(labels=labels, fontsize='large')
plt.show()
```
