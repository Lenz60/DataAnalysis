# Pie Chart


```python
df_fuel_engine_type = pd.read_csv('dataset/pie_chart_data.csv')
```

```python
sns.set_palette('muted')
```

```python
plt.figure(figsize=(9, 6))
plt.pie(df_fuel_engine_type['Number of Cars'], 
    labels=df_fuel_engine_type['Engine Fuel Type'], 
    autopct='%.2f%%',
    textprops={
        'size': 'medium',
        'fontweight': 'bold',
        
    }, explode=(0,0.1,0,0))
plt.gca().set_aspect('equal')


# Rotate the percentage labels
for text in plt.gca().texts:
    if '%' in text.get_text():
        text.set_rotation(45)
        text.set_color('white')

total = sum(df_fuel_engine_type['Number of Cars'])
percentages = [f'{(value / total) * 100:.2f}%' for value in df_fuel_engine_type['Number of Cars']]
# Check if the percentage is greater than a specified value and add it to the plot text
threshold = 30.0  # Example threshold value
plt.text(0,-1.2, 'Fuel used with more than 30% of total cars', ha='center', va='center', fontsize=12, fontweight='bold')
for i, percentage in enumerate(percentages):
    if float(percentage.strip('%')) > threshold:
        plt.text(0, -1.4 - (i * 0.1), f'{df_fuel_engine_type["Engine Fuel Type"][i]}: {percentage}', ha='center', va='center')

plt.legend(loc='upper left', fontsize='small')
plt.title('Engine Fuel Type Distribution', fontsize=15, fontweight='bold')
plt.savefig('plots/Engine Fuel Type Distribution.png')
plt.show()
```
