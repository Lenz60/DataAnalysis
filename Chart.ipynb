```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
df_used_cars = pd.read_csv('dataset/bar_chart_data.csv')
```

```python
df_used_cars
```

```python
plt.figure(figsize = (9, 6))
plt.bar(x = df_used_cars['Brand'], height = df_used_cars['Cars Listings'], color = 'mediumseagreen')
plt.xticks(rotation = 45, fontsize = 12)
plt.yticks(fontsize = 12)
plt.title('Number of Cars Listed by Brand', fontsize = 15, fontweight = 'bold')
plt.ylabel('Number of Cars Listed', fontsize = 12)
plt.savefig('plots/Car Listing by Brands.png')
plt.show()
```

```python
df_homework = pd.read_csv('dataset/bar_chart_homework_data.csv')
```

```python
df_homework
```

```python
df_homework_sorted = df_homework.sort_values(ascending=True, by='Cities')
plt.figure(figsize = (9, 6))
bar_asc = plt.bar(x=df_homework_sorted['Cities'], height=df_homework_sorted['Frequency'],
                  color='royalblue', edgecolor='black', linewidth=1)

# Highlight the fourth bar
bar_asc[3].set_color('darkorange')
bar_asc[3].set_linewidth(1)
bar_asc[3].set_edgecolor('black')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('Number of Cities Population', fontsize=15, fontweight='bold')
plt.ylabel('Number of Population', fontsize=12)
plt.savefig('plots/City Population_asc.png')
plt.show()
```
